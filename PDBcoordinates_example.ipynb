# Over all atom models

1. Read-in one pdb AA structure
2. Compute SASA
3. Compare to `Single_conf/AA_ChainA/resarea_SASA.xvg`

##### Numbers match for AA single chain

```python
import ccmap
from pypstruct import parseFilePDB
from pcmap.sasa.atom_map import atom_default_radii
import re

#pdb_folder = "../ccmap/tests/structures/"
#pdbREC = parseFilePDB(filename=f"{pdb_folder}/small_peptide_noH_1model.pdb")

def pdb_file_dictorize_noH(pdb_file_path, segID):
    pdbREC = parseFilePDB(filename=pdb_file_path)

    pdbDictREC = pdbREC.chain(segID).atomDictorize
    noH_dict = {
        "x" : [],
        "y" : [],
        "z" : [],
        "seqRes" : [],
                        "chainID" : [],
                        "resName" : [],
                        "name" : []
    }

    for x,y,z,seqRes,chainID,resName, name in zip(pdbDictREC["x"], pdbDictREC["y"], pdbDictREC["z"],\
                                        pdbDictREC["seqRes"], pdbDictREC["chainID"],\
                                        pdbDictREC["resName"], pdbDictREC["name"]
                                        ):
    #    if name.startswith("H"):
    #        continue
        if re.match(r'^[0-9]*H', name):
            continue
        noH_dict["x"].append(x)
        noH_dict["y"].append(y)
        noH_dict["z"].append(z)
        noH_dict["seqRes"].append(seqRes)
        noH_dict["chainID"].append(chainID)    
        noH_dict["resName"].append(resName)
        noH_dict["name"].append(name)
    return noH_dict
```

```python
single_conf_aa = pdb_file_dictorize_noH("./Single_conf/AA.pdb", "A")
print(single_conf_aa['name'])
len(single_conf_aa["chainID"])
```

```python
aa_asa = ccmap.sasa(single_conf_aa, atom_default_radii, probe=1.91)
aa_asa
```

```python
aa_asa_many = ccmap.sasa_list([single_conf_aa for _ in range(0,20)], atom_default_radii, probe=1.91)
```
